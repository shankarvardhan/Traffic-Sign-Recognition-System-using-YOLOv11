```python
from ultralytics import YOLO

# load a pretrained YOLOv8 model (you can try n, s, m)
model = YOLO("yolov11m.pt")   

results = model.train(
    data=data_yaml_path,
    epochs=90,          # increase if underfitting
    imgsz=640,
    batch=16,           # tune based on your GPU
    patience=20,        # early stopping
    lr0=0.001,          # base lr
    optimizer="SGD",    # or "AdamW"
    cos_lr=True,
    augment=True,
    name="traffic_signs_lights_v2"
)
```

```python
# FULL TEST SCRIPT (run this one cell in Colab)

# 1. Install dependencies
!pip install ultralytics

# 2. Imports
from ultralytics import YOLO
from google.colab import files
import os
from IPython.display import Image, display

# 3. Load your trained model
model_path = "/content/best (4).pt"  # change if needed
model = YOLO(model_path)
print("Model loaded:", model_path)

# 4. Create results folder
results_dir = "/content/results"
os.makedirs(results_dir, exist_ok=True)

# 5. Upload images from local machine
print("Please upload one or more images...")
uploaded = files.upload()   # shows upload button

input_paths = []
for filename, data in uploaded.items():
    in_path = os.path.join("/content", filename)
    with open(in_path, "wb") as f:
        f.write(data)
    input_paths.append(in_path)

print("Uploaded files:", input_paths)

# 6. Run YOLO detection on uploaded images
pred_results = model.predict(
    source=input_paths,      # list of uploaded image paths
    imgsz=640,               # can change to 768/960 later
    conf=0.25,
    save=True,               # save images with boxes
    project=results_dir,     # base folder: /content/results
    name="detect",           # subfolder: /content/results/detect
    exist_ok=True
)

out_dir = os.path.join(results_dir, "detect")
print("Predictions saved under:", out_dir)

# 7. Display result images
for file in sorted(os.listdir(out_dir)):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        img_path = os.path.join(out_dir, file)
        print("Result:", file)
        display(Image(img_path))
```
